```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import mode
from scipy.stats import probplot

# Step 1: Load the dataset
bm = pd.read_csv("datatraining.txt")

# Step 2: Summary statistics
print(bm.describe())

# Step 3: Find the mode for each column
def find_mode(column):
    return mode(column)

print("Mode of Temperature:", find_mode(bm['Temperature']))
print("Mode of Humidity:", find_mode(bm['Humidity']))
print("Mode of Light:", find_mode(bm['Light']))
print("Mode of CO2:", find_mode(bm['CO2']))
print("Mode of HumidityRatio:", find_mode(bm['HumidityRatio']))

# Step 4: Boxplots and histograms
columns_to_plot = ['Temperature', 'Humidity', 'Light', 'CO2', 'HumidityRatio']

for col in columns_to_plot:
    # Boxplot
    sns.boxplot(data=bm[col], color='steelblue')
    plt.title(f'Boxplot of {col}')
    plt.show()
    
    # Histogram
    plt.hist(bm[col], bins=20, color='steelblue', alpha=0.7)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Step 5: Q-Q Plots
for col in columns_to_plot:
    probplot(bm[col], dist="norm", plot=plt)
    plt.title(f'Q-Q Plot for "{col}" Normality')
    plt.show()
```
