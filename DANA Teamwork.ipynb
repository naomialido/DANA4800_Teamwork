Date
Type: Categorical
Subtype: Nominal
Reason: Even though it looks numeric, it's just a label. Dates represent unique points in time with no measurable quantity or ranking.
Not discrete or continuous, as it's not a number we calculate with directly (unless we extract parts like hour, weekday, etc.).

Temperature
Type: Numerical
Subtype: Interval
Nature: Continuous
Reason: Temperature has meaningful differences between values (e.g., 25°C is 5 degrees warmer than 20°C), but it doesn’t have a true zero — 0°C doesn’t mean “no temperature.” This makes it interval, not ratio.

Humidity
Type: Numerical
Subtype: Ratio
Nature: Continuous
Reason: Humidity is a measurable amount, with a true zero — 0% means no humidity.

Light
Type: Numerical
Subtype: Ratio
Nature: Continuous
Reason: Light intensity has a true zero (total darkness) and values can be compared meaningfully. For example, 800 lux is twice as bright as 400 lux.

CO2
Type: Numerical
Subtype: Ratio
Nature: Continuous
Reason: CO2 concentration is measurable with a meaningful zero — no CO2 means 0 ppm. Ratios and differences matter.

HumidityRatio
Type: Numerical
Subtype: Ratio
Nature: Continuous
Reason: This is a physical ratio of water vapor to air with a true zero (dry air). It’s also measured continuously.

Occupancy
Type: Categorical
Subtype: Binary (Nominal)
Nature: Discrete
Reason: There are only two categories: 0 = not occupied, 1 = occupied. These are labels, not measurable quantities, so it's categorical, and because there are only two possible values, it’s binary. It’s also discrete because we can count people but not have 0.5 of an occupancy.

```python
# Function to classify numerical columns as discrete or categorical based on unique values
```

```python
def classify_numerical_column(column, threshold_discrete=15, threshold_categorical=10):
    unique_values = column.nunique()
    if unique_values <= threshold_categorical:
        return 'Categorical'
    elif unique_values <= threshold_discrete:
        return 'Discrete Numerical'
    else:
        return 'Continuous Numerical'
```

```python
# Apply the classification to all numerical columns in training dataset
```

```python
df = training 
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

classification_results = {
    col: {
        'unique_values': df[col].nunique(),
        'classification': classify_numerical_column(df[col])
    } for col in numerical_columns
}

classification_results
```

If a column has 2–5 unique values, and values represent categories, it is likely categorical 
If a column has 5–15 unique values, and those values represent actual quantities => it can be treated as discrete numerical.
If it has many (e.g., >15 or 20) unique values, it's almost always numerical — either discrete (counts) or continuous (measurements).

According to the rule of thumb:
Anything with >15 unique values is treated as continuous.
Occupancy with just 2 values is correctly identified as a categorical variable.

```python
from scipy.stats import pearsonr, spearmanr, kruskal
from sklearn.feature_selection import mutual_info_classif
```

```python
#For the training dataset
```

```python

y = df_viz['Occupancy']
numerical_features = ['Temperature', 'Humidity', 'Light', 'CO2', 'HumidityRatio']
correlation_results = {}

for feature in numerical_features:
    pearson_corr, _ = pearsonr(df_viz[feature], y)
    spearman_corr, _ = spearmanr(df_viz[feature], y)
    kruskal_stat, kruskal_p = kruskal(*[df_viz[df_viz['Occupancy'] == val][feature] for val in df_viz['Occupancy'].unique()])
    correlation_results[feature] = {
        'Pearson': pearson_corr,
        'Spearman': spearman_corr,
        'Kruskal-Wallis H': kruskal_stat,
        'Kruskal-Wallis p': kruskal_p
    }
correlation_results
```

```python
# Mutual Information (non-linear dependencies)
```

```python
X = df[numerical_features]
mi_scores = mutual_info_classif(X, y, discrete_features=False, random_state=42)
for idx, feature in enumerate(numerical_features):
    correlation_results[feature]['Mutual Information'] = mi_scores[idx]

import pandas as pd
correlation_df = pd.DataFrame(correlation_results).T.round(3)
correlation_df
```

Pearson: Measures linear correlation. High for Light (0.91), CO2, and Temperature.
Spearman: Measures monotonic (rank-based) correlation. Also strong for Light and CO2.
Kruskal-Wallis: A non-parametric test for difference between groups. All p-values are 0.0, meaning statistically significant differences across occupancy classes.
Mutual Information: Captures any kind of dependency, including non-linear. Light, CO2, and Temperature again have the highest scores.

Both linear and non-linear relationships:
Light and CO2 have strong linear and non-linear relationships with Occupancy.
Humidity has weak correlation, regardless of method.
Mutual Information adds insight beyond Pearson/Spearman when relationships aren’t purely linear.

```python
from scipy.stats import pearsonr, spearmanr, kruskal
from sklearn.feature_selection import mutual_info_classif

numerical_features = ['Temperature', 'Humidity', 'Light', 'CO2', 'HumidityRatio']
X_test1 = df_test1[numerical_features]
y_test1 = df_test1['Occupancy']
```

```python
# Pearson and Spearman
```

```python
pearson_spearman_test1 = {
    feature: {
        'Pearson': pearsonr(df_test1[feature], y_test1)[0],
        'Spearman': spearmanr(df_test1[feature], y_test1)[0]
    }
    for feature in numerical_features
}
```

```python
# Kruskal-Wallis
```

```python
kruskal_test1 = {
    feature: {
        'Kruskal-Wallis H': kruskal(
            *[df_test1[df_test1['Occupancy'] == val][feature] for val in sorted(df_test1['Occupancy'].unique())]
        )[0],
        'Kruskal-Wallis p': kruskal(
            *[df_test1[df_test1['Occupancy'] == val][feature] for val in sorted(df_test1['Occupancy'].unique())]
        )[1]
    }
    for feature in numerical_features
}
```

```python
# Mutual Information
```

```python
mi_test1 = mutual_info_classif(X_test1, y_test1, discrete_features=False, random_state=42)
mi_dict_test1 = {feature: mi for feature, mi in zip(numerical_features, mi_test1)}

results_test1 = []
for feature in numerical_features:
    row = {
        'Pearson': round(pearson_spearman_test1[feature]['Pearson'], 3),
        'Spearman': round(pearson_spearman_test1[feature]['Spearman'], 3),
        'Kruskal-Wallis H': round(kruskal_test1[feature]['Kruskal-Wallis H'], 3),
        'Kruskal-Wallis p': round(kruskal_test1[feature]['Kruskal-Wallis p'], 4),
        'Mutual Information': round(mi_dict_test1[feature], 3)
    }
    results_test1.append(row)

df_results_test1 = pd.DataFrame(results_test1, index=numerical_features)
df_results_test1
```

Light: Strongest indicator of occupancy (Pearson 0.93, MI 0.60) — room lights on when people are present.
CO2: Strong correlation (Pearson 0.77) — people emit CO2, so presence increases levels.
Temperature: Moderate correlation (Pearson 0.71) — warmer during presence, but some overlap.
HumidityRatio: Decent predictor (MI 0.35) — affected by people’s respiration.
Humidity: Weakest — low correlation, doesn’t separate occupied/unoccupied well.

both linear and non-linear correlations.
Light and CO2 are the most consistent and strongest features across all tests.
